## Question 2: 

Given the soil profile shown in the figure below, (1) calculate the transfer function relating ground surface to the bedrock outcrop [i.e., $ùêπ_2(ùëì)$] and (2) plot out the amplification factor [i.e., $|ùêπ_2(ùëì)|$ vs. ùëì for ùëì = 0 to 25 Hz].

### Solution:   

1. Compute the transfer function that relates the ground surface (output) motion to the bedrock (input) motion. From equation (1), the transfer function for the case of damped soil is complex valued. 

$$ F_2(\omega) = \frac{1}{\cos k(1 - i\xi)H} = \frac{1}{\cos\left[\frac{\omega H}{v_s}(1 + i\xi)\right]} \tag{1}$$


```python
# Package import
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# Constants
H = 540  # Height in feet
vs= 1500  # Shear wave velocity in ft/s
xi = 0.05  # Damping ratio

# Frequency range (0 to 25 Hz)
frequencies = np.linspace(0, 25, 500)  # 500 points between 0 and 25 Hz
omega = 2 * np.pi * frequencies  # Angular frequency (rad/s)

# Calculate F2(omega)
F2 = 1 / np.cos((omega * H / vs) * (1 + 1j * xi))

# Print the first few F2 values for verification
print("F2 values (first 10):")
print(F2[:10])
```

2. The modulus of the transfer function is calculated based equation (2). Plot of $|ùêπ_2(ùëì)|$ vs. ùëì for ùëì = 0 to 25 Hz] is shown in the figure below. The shape of the transfer function indicates that significant amplification will occur at several natural frequencies.

$$ |F_2(\omega)| \approx \frac{1}{\sqrt{\cos^2 kH + (\xi kH)^2}} = \frac{1}{\sqrt{\cos^2 \left(\frac{\omega H}{v_s}\right) + \left[\xi \left(\frac{\omega H}{v_s}\right)\right]^2}} \tag{2}$$


```python
# Plot F_2(\omega) vs Frequency

# Magnitude of F2
F2_magnitude = np.abs(F2)

# Plot the amplification factor |F2(f)|
plt.figure(figsize=(12, 4))
plt.plot(frequencies, F2_magnitude, )
plt.title("Amplification Factor |F2(f)| vs Frequency")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplification Factor")
plt.ylim(0, 14)
plt.grid()
plt.show()
```
